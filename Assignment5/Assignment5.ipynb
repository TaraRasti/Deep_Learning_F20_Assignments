```
img1t=torch.from_numpy(img1)
img2t=torch.from_numpy(img2)
img3t=torch.from_numpy(img3)
```

```
x=img1t.permute(2,0,1)
y=img2t.permute(2,0,1)
z=img3t.permute(2,0,1)
```

```
transformations = transforms.Compose([transforms.ToPILImage(),
                                      transforms.Resize((256,256)),
                                      transforms.ToTensor()])
```

```
X=transformations(x)
Y=transformations(y)
Z=transformations(z)
```

```
print(X.shape)
print(type(X))
print(Y.shape)
print(type(Y))
print(Z.shape)
print(type(Z))
```
